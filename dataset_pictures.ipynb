```python
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches

# Set up the figure
fig, (ax1, ax2, ax3) = plt.subplots(1, 3, figsize=(15, 5))

# Data for pie charts
# 1. Camera angles
camera_angles_labels = ['Low-angle view', 'Mid-angle view', 'High-angle view', 'Bird\'s eye view']
camera_angles_sizes = [12.2, 60.1, 23.9, 3.8]
camera_angles_colors = ['#f8d0c2', '#4a9294', '#f08e79', '#a8d1c8']

# 2. Scenery/Weather
scenery_labels = ['Outdoor(Sunny)', 'Indoor', 'Outdoor(cloudy/rainy)', 'Outdoor(Night)']
scenery_sizes = [29.8, 26.5, 29.4, 14.3]
scenery_colors = ['#f8d0c2', '#4a9294', '#f08e79', '#a8d1c8']

# 3. Light condition
light_labels = ['Good light', 'Poor light']
light_sizes = [82.4, 17.6]
light_colors = ['#f8d0c2', '#4a9294']

# First pie chart - Camera angles
wedges1, _, autotexts1 = ax1.pie(camera_angles_sizes, colors=camera_angles_colors, 
                                autopct='%1.1f%%', startangle=90)
ax1.set_title('Camera angles', fontsize=16)
plt.setp(autotexts1, size=12, weight='bold')

# Create legend for first chart
handles1 = [mpatches.Patch(color=color) for color in camera_angles_colors]
legend1 = ax1.legend(handles1, camera_angles_labels, loc='center left', 
                    bbox_to_anchor=(-0.2, 0.5), frameon=True, fontsize=16)
legend1.get_frame().set_linewidth(0.5)

# Second pie chart - Scenery/Weather
wedges2, _, autotexts2 = ax2.pie(scenery_sizes, colors=scenery_colors, 
                                autopct='%1.1f%%', startangle=90)
ax2.set_title('Scenery / Weather', fontsize=16)
plt.setp(autotexts2, size=14, weight='bold')

# Create legend for second chart
handles2 = [mpatches.Patch(color=color) for color in scenery_colors]
legend2 = ax2.legend(handles2, scenery_labels, loc='center left', 
                    bbox_to_anchor=(-0.2, 0.5), frameon=True, fontsize=16)
legend2.get_frame().set_linewidth(0.5)

# Third pie chart - Light condition
wedges3, _, autotexts3 = ax3.pie(light_sizes, colors=light_colors, 
                                autopct='%1.1f%%', startangle=90)
ax3.set_title('Light condition', fontsize=16)
plt.setp(autotexts3, size=14, weight='bold')

# Create legend for third chart
handles3 = [mpatches.Patch(color=color) for color in light_colors]
legend3 = ax3.legend(handles3, light_labels, loc='center left', 
                    bbox_to_anchor=(-0.2, 0.5), frameon=True, fontsize=16)
legend3.get_frame().set_linewidth(0.5)

# Ensure pie charts are circles
ax1.axis('equal')
ax2.axis('equal')
ax3.axis('equal')

# Adjust spacing between subplots
plt.tight_layout()
plt.subplots_adjust(wspace=0.5)

plt.show()
```
